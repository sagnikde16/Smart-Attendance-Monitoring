```python
# !pip install ultralytics opencv-python

# import os
# import cv2
# from ultralytics import YOLO

BASE_DIR = "/content/drive/MyDrive/smartAttendence"
# /content/drive/MyDrive/smartAttendence/models/yolo/yolov8l-face_2.pt

YOLO_MODEL_PATH = os.path.join(BASE_DIR, "models/yolo/yolov8l-face_2.pt")
VIDEO_PATH = os.path.join(BASE_DIR, "videos/30_07_2025.mp4")
# /content/drive/MyDrive/smartAttendence/videos/30_07_2025 (online-video-cutter.com).mp4
FRAMES_DIR = os.path.join(BASE_DIR, "output/frames")
FACES_DIR = os.path.join(BASE_DIR, "output/faces")

os.makedirs(FRAMES_DIR, exist_ok=True)
os.makedirs(FACES_DIR, exist_ok=True)

print("Loading YOLO...")
model = YOLO(YOLO_MODEL_PATH)

cap = cv2.VideoCapture(VIDEO_PATH)

frame_id = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1

    cv2.imwrite(os.path.join(FRAMES_DIR, f"frame_{frame_id}.jpg"), frame)

    results = model(frame, conf=0.5)
    face_id = 0

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            face = frame[y1:y2, x1:x2]
            if face.size == 0:
                continue

            face_id += 1
            cv2.imwrite(os.path.join(FACES_DIR, f"frame_{frame_id}_face_{face_id}.jpg"), face)

    print(f"Processed frame {frame_id}")

cap.release()
print("Done.")
```

```python
# import os
# import cv2
# import numpy as np

BASE_DIR = "/content/drive/MyDrive/smartAttendence"

FACES_DIR = os.path.join(BASE_DIR, "output/faces")
FILTERED_DIR = os.path.join(BASE_DIR, "output/faces_filtered")

os.makedirs(FILTERED_DIR, exist_ok=True)

def is_blurry(image, threshold=100.0):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lap_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return lap_var < threshold

min_face_size = 40   # minimum width or height

kept = 0
removed_blur = 0
removed_small = 0

for fname in os.listdir(FACES_DIR):
    fpath = os.path.join(FACES_DIR, fname)
    img = cv2.imread(fpath)

    if img is None:
        continue

    h, w, _ = img.shape

    # Size filter
    if h < min_face_size or w < min_face_size:
        removed_small += 1
        continue

    # Blur filter
    if is_blurry(img, threshold=100.0):
        removed_blur += 1
        continue

    cv2.imwrite(os.path.join(FILTERED_DIR, fname), img)
    kept += 1

print("Filtering done!")
print("Kept faces:", kept)
print("Removed (small):", removed_small)
print("Removed (blurry):", removed_blur)
```

```python
import os
import json
from deepface import DeepFace

BASE_DIR = "/content/drive/MyDrive/smartAttendence"
FACES_FILTERED_DIR = os.path.join(BASE_DIR, "output/faces_filtered")
EMB_DIR = os.path.join(BASE_DIR, "output/embeddings")
os.makedirs(EMB_DIR, exist_ok=True)

embeddings = {}

files = sorted(os.listdir(FACES_FILTERED_DIR))
print("Total faces:", len(files))

for i, fname in enumerate(files):
    fpath = os.path.join(FACES_FILTERED_DIR, fname)
    try:
        rep = DeepFace.represent(
            img_path=fpath,
            model_name="VGG-Face",
            detector_backend="skip",   # we already cropped faces
            enforce_detection=False
        )
        embeddings[fname] = rep[0]["embedding"]
    except Exception as e:
        print("Error on", fname, e)

    if (i+1) % 50 == 0:
        print(f"Processed {i+1}/{len(files)} faces")

# Save embeddings
emb_path = os.path.join(EMB_DIR, "embeddings.json")
with open(emb_path, "w") as f:
    json.dump(embeddings, f)

print("Embeddings saved at:", emb_path)
```

```python
import os
import json
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import normalize

BASE_DIR = "/content/drive/MyDrive/smartAttendence"
EMB_PATH = os.path.join(BASE_DIR, "output/embeddings/embeddings.json")
CLUSTERS_DIR = os.path.join(BASE_DIR, "output/clusters")
os.makedirs(CLUSTERS_DIR, exist_ok=True)

# Load embeddings
with open(EMB_PATH, "r") as f:
    emb_dict = json.load(f)

filenames = list(emb_dict.keys())
X = np.array(list(emb_dict.values()), dtype=np.float32)

# Normalize embeddings (important for cosine-like behavior)
X = normalize(X)

print("Total embeddings:", X.shape)

# DBSCAN clustering
dbscan = DBSCAN(
    eps=0.35,        # tune this (0.3‚Äì0.6 typical)
    min_samples=5,  # minimum faces to form a student
    metric="euclidean"
)

labels = dbscan.fit_predict(X)

# Save cluster-wise file lists
clusters = {}
for fname, label in zip(filenames, labels):
    if label == -1:
        continue  # noise
    clusters.setdefault(int(label), []).append(fname)

print("Total clusters (students):", len(clusters))

# Save clusters to JSON
clusters_path = os.path.join(CLUSTERS_DIR, "clusters.json")
with open(clusters_path, "w") as f:
    json.dump(clusters, f, indent=2)

print("Clusters saved at:", clusters_path)
```

```python
import shutil

FACES_FILTERED_DIR = os.path.join(BASE_DIR, "output/faces_filtered")
CLUSTERS_VIS_DIR = os.path.join(BASE_DIR, "output/clusters_vis")
os.makedirs(CLUSTERS_VIS_DIR, exist_ok=True)

for cid, files in clusters.items():
    cluster_folder = os.path.join(CLUSTERS_VIS_DIR, f"cluster_{cid}")
    os.makedirs(cluster_folder, exist_ok=True)

    for fname in files:
        src = os.path.join(FACES_FILTERED_DIR, fname)
        dst = os.path.join(cluster_folder, fname)
        shutil.copy(src, dst)

print("Visual clusters created at:", CLUSTERS_VIS_DIR)
```

```python
import os
import json
from deepface import DeepFace

BASE_DIR = "/content/drive/MyDrive/smartAttendence"
PASSPORT_DIR = os.path.join(BASE_DIR, "output/passport_photos")
PASSPORT_EMB_PATH = os.path.join(BASE_DIR, "output/passport_embeddings.json")

passport_embeddings = {}

for fname in os.listdir(PASSPORT_DIR):
    fpath = os.path.join(PASSPORT_DIR, fname)
    roll_no = os.path.splitext(fname)[0]

    try:
        rep = DeepFace.represent(
            img_path=fpath,
            model_name="VGG-Face",
            detector_backend="skip",
            enforce_detection=False
        )
        passport_embeddings[roll_no] = rep[0]["embedding"]
    except Exception as e:
        print("Error processing passport photo:", fname, e)

with open(PASSPORT_EMB_PATH, "w") as f:
    json.dump(passport_embeddings, f)

print("Passport embeddings saved at:", PASSPORT_EMB_PATH)
```

```python
import json
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

BASE_DIR = "/content/drive/MyDrive/smartAttendence"

# Load clusters
with open(os.path.join(BASE_DIR, "output/clusters/clusters.json")) as f:
    clusters = json.load(f)

# Load face embeddings
with open(os.path.join(BASE_DIR, "output/embeddings/embeddings.json")) as f:
    face_emb = json.load(f)

# Load passport embeddings
with open(os.path.join(BASE_DIR, "output/passport_embeddings.json")) as f:
    passport_emb = json.load(f)

cluster_to_identity = {}

for cid, files in clusters.items():
    # Compute cluster mean embedding
    embs = [face_emb[fname] for fname in files]
    cluster_mean = np.mean(np.array(embs), axis=0).reshape(1, -1)

    best_roll = None
    best_score = -1

    for roll_no, ref_emb in passport_emb.items():
        score = cosine_similarity(cluster_mean, np.array(ref_emb).reshape(1, -1))[0][0]
        if score > best_score:
            best_score = score
            best_roll = roll_no

    cluster_to_identity[cid] = {
        "roll_no": best_roll,
        "similarity": float(best_score)
    }

print("Cluster ‚Üí Identity mapping:")
cluster_to_identity
```

```python
import pandas as pd

attendance = []

for cid, info in cluster_to_identity.items():
    attendance.append({
        "Cluster_ID": cid,
        "Roll_No": info["roll_no"],
        "Similarity": info["similarity"],
        "Status": "Present"
    })

df = pd.DataFrame(attendance)
ATT_PATH = os.path.join(BASE_DIR, "output/attendance.xlsx")
df.to_excel(ATT_PATH, index=False)

print("Attendance saved at:", ATT_PATH)
df
```
